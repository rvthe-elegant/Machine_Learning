```python
#importing the libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
#loading the crx.data dataset
data=pd.read_csv('data.csv')
```

```python
data
```

```python
#Task 1 : Replacing ? with NaN
data=data.replace('?',np.nan)
data
```

```python
#Task 2: Renaming columns to easier notations
data.rename(columns = {'property type(apartment, villa)':'prop_type'}, inplace = True)
data.rename(columns = {'property sale(new/resale)':'prop_sale'}, inplace = True)
data.head()
```

```python
#Task 3: printing unique values of mutiple attributes
print(data['prop_type'].unique())
print('\n')
print(data['prop_sale'].unique())
```

```python
data['NoOfBedroom'] = data['NoOfBedroom'].replace(np.nan, 0)
data
```

```python
data.info()
```

```python
data['price']=data['price'].astype('double')
data['NoOfBedroom']=data['NoOfBedroom'].astype('double')
```

```python
#Tak4: Plotting graph betweeen 'NoOfBedroom' and 'price'
X = (data['NoOfBedroom'])
Y = (data['price']) 
plt.xlabel('NoOfBedroom')
plt.ylabel('price')
plt.plot(X,Y)
plt.title('NoOfBedroom vs. Price')
```

```python
#Tak 5: Plotting  bar graph betweeen 'Place' and 'price'
groupedby_places = data.groupby('Place')[['price']].mean()
print(groupedby_places.sort_values(by = ['price']))
plt.figure(figsize=(9,6))
plt.bar(x=data['Place'],height=data['price'])
plt.xticks(rotation=45)
```
